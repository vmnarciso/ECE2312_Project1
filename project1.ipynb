```python
# import necessary modules
import sounddevice as sd
from scipy.io import wavfile
from scipy import signal 
from scipy.io.wavfile import write
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
```

```python
# SENTENCE 1

# Define Sampling Rate or Frequency in Hz
sr = 44100

# Record duration in seconds
duration = 5
```

```python
# Start audio recording
recording = sd.rec(int(duration*sr), samplerate=sr, channels=1) # we will record with a  mono or stereo channel microphone

# Record audio for the given duration
print("recording...............")
sd.wait()

# Write it to a file
write("The quick brown fox jumps over the lazy dog.wav",sr,recording)
```

```python
# Look at the discrete number array we got from the audio
x, sr = librosa.load('The quick brown fox jumps over the lazy dog.wav')
print("recording shape", x.shape)
print("sampling rate", sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr, color='r')
plt.title("The quick brown fox jumps over the lazy dog")
plt.xlabel("Time (s)")
plt.ylabel("Frequency") 
```

```python
freq = librosa.stft(x)
freqdb = librosa.amplitude_to_db(abs(freq), ref=8000)
plt.figure(figsize=(10, 5))
librosa.display.specshow(freqdb, sr=sr, x_axis='time', y_axis='hz')
plt.title("The quick brown fox jumps over the lazy dog")
plt.show()
```

```python
# Play the audio
ipd.Audio(x, rate=sr) # load a NumPy array
```
